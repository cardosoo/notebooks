```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from fitutils import linfitxy
```

```python
x = np.arange(5) + rd.normal(0, 0.1, 5)
y = 2 * np.arange(5) + 12 + rd.normal(0, 0.1, 5)
sx = np.ones(5) * 0.4
sy = np.ones(5) * 1
print(x, y)
```

```python
res = linfitxy(x, y, sx, sy, plot=True)
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('fig.png')
plt.show()
```
